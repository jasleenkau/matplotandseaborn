```python

#load the data
data=pd.read_csv("/content/drive/MyDrive/dataset/US_Accidents_Dec21_updated.csv")
data
```

```python

#1. Load the dataset into a pandas DataFrame and check for missing values.
data.isnull()
data.dropna(inplace= True)
data
```

```python
#Identify the top 10 states with the highest number of accidents and
#create a bar chart using matplotlib to visualize the data
```

```python
top_10_state=data['State'].value_counts().head(10)
plt.bar(top_10_state.index, top_10_state.values)
plt.xlabel('state')
plt.ylabel('Number of accients')
plt.title('Top 10 highest number of accients')
plt.show()
```

```python
#3. Create a scatter plot of the longitude and latitude columns to show the location of accidents.
longitude = data['Start_Lng']
latitude = data['Start_Lat']
plt.scatter(longitude, latitude, s=1, alpha=0.5)  # Adjust the size and transparency of the dots as needed
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Accident Locations')
plt.show()
```

```python
#4.Create a pie chart to show the distribution of accidents by severity (use the 'Severity' column)
severity_counts = data['Severity'].value_counts()
labels = severity_counts.index
colors = ['red', 'orange', 'yellow', 'green']
explode = (0.1, 0, 0, 0)  # To explode the first slice (optional)
plt.pie(severity_counts, labels=labels, colors=colors, explode=explode, autopct='%1.1f%%')
plt.title('Distribution of Accidents by Severity')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle
plt.show()

```

```python
#5.Identify the top 10 cities with the highest number of accidents and
#create a horizontal bar chart using matplotlib to visualize the data
top_10_cities=data['City'].value_counts().head(10)
plt.bar(top_10_cities.index, top_10_cities.values)
plt.xlabel("Cities")
plt.ylabel("Number of accidents")
plt.title("Top 10 cities with highest number of accident")
plt.show()
```

```python
#Create a line chart to show the trend of accidents over time (use the 'Start_Time' column)
data['date']=pd.to_datetime(data['Start_Time']).dt.date
data[['Start_Time', 'date']]
accident_over_time=data['date'].value_counts().sort_index()
plt.plot(accident_over_time.index,accident_over_time.values)
plt.xlabel('Date')
plt.ylabel('Number of accidents')
plt.title('Trend of accidents over time')
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
#prepare and load
Titanic=sns.load_dataset("titanic")
#1. Identify missing values: Check for any missing values in the Age column.
Titanic['age'].isnull().sum()

# 2.Fill in missing values: If there are missing values in the Age column,
#use appropriate techniques such as imputation or interpolation to fill in the missing values.
Titanic['age'].fillna("unknown")
#deleting the missing values
Titanic.dropna(inplace=True)
Titanic

#3.Inconsistent formats: Check if there are any inconsistencies in the formatting of the gender column.
#For example, if the gender is encoded as "male" and "female", ensure that it is consistent throughout the dataset.


# Convert inconsistent gender values to a consistent format
Titanic['sex'] =Titanic['sex'].replace({'male': 'Male', 'female': 'Female'})
Titanic



# Convert the data type of the Age column to numeric (float or integer)
Titanic['age'] = pd.to_numeric(Titanic['age']).head(10)

# Create a plot using Seaborn
sns.catplot(x='age', y='survived', hue='sex', data=Titanic, kind='bar')

# Set the title and labels
plt.title('Top 10 Survival by Age and Gender')
plt.xlabel('Age')
plt.ylabel('Survived')

# Display the plot
plt.show()







```

```python


# Create a box plot to visualize the distribution of age by survival status
sns.boxplot(x='survived', y='age', data=Titanic)
plt.title('Age Distribution by Survival Status')
plt.xlabel('Survived')
plt.ylabel('Age')
plt.show()

# Create a swarm plot to visualize the relationship between age, gender, and survival status
sns.swarmplot(x='survived', y='age', hue='sex', data=Titanic)
plt.title('Age and Gender by Survival Status')
plt.xlabel('Survived')
plt.ylabel('Age')
plt.legend(title='Sex', loc='upper right')
plt.show()
```

```python
#Create data visualizations: Use Seaborn to create appropriate visualizations such as bar charts, count plots,
#or heatmaps to explore the relationship between age, gender, and survival status of the passengers on the Titanic


# Create a count plot to visualize the survival status by gender
sns.countplot(x='survived', hue='sex', data=Titanic)
plt.title('Survival Status by Gender')
plt.xlabel('Survived')
plt.ylabel('Count')
plt.show()

# Create a box plot to visualize the age distribution by survival status
sns.boxplot(x='survived', y='age', data=Titanic)
plt.title('Age Distribution by Survival Status')
plt.xlabel('Survived')
plt.ylabel('Age')
plt.show()

# Create a heatmap to visualize the correlation between variables
correlation = Titanic[['survived', 'age']].corr()
sns.heatmap(correlation, annot=True, cmap='coolwarm')
plt.title('Correlation between Survival and Age')
plt.show()
```

```python
#Interpretation: Based on the visualizations, draw conclusions and insights about
# the relationship between age, gender, and survival status of the passengers on the Titanic



# Filter the dataset to include only relevant columns
data = Titanic[['age', 'sex', 'survived']]

# Replace numeric values of 'Survived' column with corresponding labels
data['survived'] = data['survived'].map({0: 'No', 1: 'Yes'})

# Create a countplot to visualize the number of survivors by gender
sns.countplot(x='sex', hue='survived', data=data)
plt.title('Survivors by Gender')
plt.show()

# Create a violin plot to visualize the distribution of age by survival status and gender
sns.violinplot(x='survived', y='age', hue='sex', data=data, split=True)
plt.title('Distribution of Age by Survival Status and Gender')
plt.show()
```

```python

#prepare and load


# 1. Load the Iris Species dataset
iris_data = sns.load_dataset('iris')

# 2. Explore the data: Use Seaborn's functions such as head(), info(),
#and describe() to explore the dataset and understand its structure, contents, and data types.
print(iris_data.head())
iris_data.info()
iris_data.describe()
#Data visualization: Create various visualizations using Seaborn's functions such as scatterplot(), boxplot(), violinplot(), pairplot(), heatmap(), etc. to visualize the relationships between different variables and gain insights
#from the data. Experiment with different types of visualizations to explore different aspects of the data.


# Scatterplot
sns.scatterplot(data=iris_data, x='sepal_length', y='sepal_width', hue='species')
plt.title('Scatterplot of Sepal Length vs Sepal Width')
plt.show()

# Boxplot
sns.boxplot(data=iris_data, x='species', y='petal_length')
plt.title('Boxplot of Petal Length by Species')
plt.show()

# Violinplot
sns.violinplot(data=iris_data, x='species', y='petal_width')
plt.title('Violinplot of Petal Width by Species')
plt.show()

# Pairplot
sns.pairplot(data=iris_data, hue='species')
plt.title('Pairplot of Iris Dataset')
plt.show()

# Heatmap
correlation_matrix = iris_data.corr()
sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation Heatmap of Iris Dataset')
plt.show()


```
